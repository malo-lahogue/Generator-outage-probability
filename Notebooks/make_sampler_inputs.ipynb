From a result file, create inputs file

```python
import pandas as pd
import numpy as np
```

```python
result_file = '/Users/malo/MIT Dropbox/Malo Lahogue/Research/Generator-outage-probability/Results/Transition_models/GAMS/SMC/GAM__SMC_fullH_GAM__results_test_ALABAMA.csv'
```

```python
results_df = pd.read_csv(result_file, parse_dates=['Datetime_UTC'])
```

```python
results_df.head()
```

```python
keep_cols = [c for c in results_df.columns if (not c.startswith('pA')) and  
                                              (not c.startswith('pD')) and 
                                              (not c.startswith('pO')) and 
                                              (not c in ['Initial_gen_state', 'Final_gen_state', 'Data_weight'])]
```

```python
inputs_df = results_df[keep_cols].copy()
inputs_df.drop_duplicates(subset=['Datetime_UTC', 'State'], inplace=True)
```

```python
inputs_df
```

```python
inputs_df.to_csv(f"../DATA/hourly/scenario_inputs_test_2022-2023.csv", index=False)
```
